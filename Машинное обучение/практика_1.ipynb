```python
sns.histplot(df['Price'], kde=True)
plt.title('Распределение Price')
plt.xlabel('Price')
plt.ylabel('Частота')
plt.show()
```

```python
sns.countplot(x='Carat Weight', data=df)
plt.title('Распределение weight')
plt.xlabel('price')
plt.ylabel('Количество')
plt.show()
```

```python
sns.boxplot(x=df['Price'])
plt.title('Boxplot chlorides')
plt.xlabel('price')
plt.show()
```

Ваши наблюдения:
Распледеление по price нормальное, carat weight распределён по разному, в осноном всё до 2кг, при этом большинство значений находятся в промежутке 1 и 1.5. Значения в boxplot сосредоточина в промежутке 25000 и 40000.
